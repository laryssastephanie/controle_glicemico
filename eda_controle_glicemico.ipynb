```python
import pandas as pd
import plotly.graph_objects as go
```

```python
data = pd.read_excel('controle_glicemico.xlsx', engine="openpyxl")
```

```python
data.head()
```

```python
fig = go.Figure()
fig.add_trace(go.Bar(x=data['Dia'],
                    y=data['Jejum'], name='Jejum'))
fig.add_trace(go.Bar(x=data['Dia'],
                    y=data['Pós almoço 2h'], name='Pós 2h almoço'))
fig.add_trace(go.Bar(x=data['Dia'],
                    y=data['Pós Jantar 2h'], name='Pós 2h Jantar'))
fig.add_hline(y=120, 
            annotation_text="Valor de referência pós prandial", 
            annotation_position="bottom right",
            line_color='tomato')
fig.add_hline(y=95, 
            annotation_text="Valor de referência Jejum", 
            annotation_position="bottom right",
            line_color='purple')
fig.update_layout(
    title='Controle Glicêmico',
    xaxis_tickfont_size=14,
    yaxis=dict(
        title='Resultado HTG',
        titlefont_size=16,
        tickfont_size=14,
    ),
    legend=dict(
        x=1,
        y=0),
    barmode='group',
    bargap=0.4, # gap between bars of adjacent location coordinates.
    bargroupgap=0.2 # gap between bars of the same location coordinate.
)
```
